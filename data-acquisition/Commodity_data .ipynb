```python
from lxml.etree import fromstring
import pandas as pd
import calendar, requests
#from indices.indices import *
#from indices.commodities import *

# set https header parameters
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36', #required 
    'referer': "https://www.investing.com/commodities/us-corn-historical-data",
    'host' : 'www.investing.com',
    'X-Requested-With' : 'XMLHttpRequest'
}

class IndiceHistoricalData():    
    
    def __init__(self, API_url):
        self.API_url = API_url

    #set https header for request
    def setHeaders(self, headers):
        self.headers = headers 
    
    #set indice data (indices.py) 
    def setFormData(self, data):
        self.data = data 

       
    def updateFrequency(self, frequency):
        self.data['interval_sec'] = frequency     
        

    def updateStartingEndingDate(self, startingDate, endingDate):
        self.data['st_date'] = startingDate     
        self.data['end_date'] = endingDate     


    def setSortColumns(self, sorting_columns):
        self.data['sort_col'] = sorting_columns

    def setSortOreder(self, sorting_order):
        self.data['sort_ord'] = sorting_order   

    def downloadData(self):
        #print(self.data)
        self.response = requests.post(self.API_url, data=self.data, headers=self.headers).content
        #parse tables with pandas - [0] probably there is only one html table in response
        self.observations = pd.read_html(self.response)[0]
        #print(type(self.observations['Date']))
        return self.observations

    def printData(self):
        print(type(self.observations['Date']))# = self.observations['Date'].replace(" ","-")
        print(self.observations)

    def saveDataCSV(self):
        #print(str(self.observations['Date']).replace(" ",""))
        self.observations['Date'] = "25-"+self.observations['Date'].str.replace(" ","-")
        self.observations['Date'] = pd.to_datetime(self.observations['Date'],errors = 'coerce')
        self.observations.to_csv('./indices/'+self.data['header']+" ("+self.data['name']+")"+'.csv', sep=',', encoding='utf-8',index=False)
        #return self.observations

lst = [GCG8, HGH8, SIH8, NGF8, ZCN9,\
         MALTR, CLN9, RBN9, HON9, SBN9,\
         KCN9, CCN9, RRN9, LEQ9, HEM9, \
         FCQ9, DCSc1, CTN9, ZWN9, ZLN9, \
         Oc1, ZSN9, HGN9, GCQ9, SIN9, PLN9, PAU9]
        

for i in lst:
    ihd = IndiceHistoricalData('https://www.investing.com/instruments/HistoricalDataAjax')
    ihd.setHeaders(headers)
    ihd.setFormData(i)
    
    ihd.updateFrequency('Monthly')
    ihd.updateStartingEndingDate('01/01/1980', '12/20/2019')
    ihd.setSortOreder('ASC')
    ihd.setSortColumns('date')
    ihd.downloadData()
    #ihd.printData()
    ihd.saveDataCSV()
```
